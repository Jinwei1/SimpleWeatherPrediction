```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, mean_absolute_error
from math import sqrt
weather_df = pd.read_csv("data/weatherHistory.csv")
weather_df.plot(x='Humidity', y='Apparent Temperature (C)',style='.')
X = weather_df['Humidity']
Y = weather_df['Apparent Temperature (C)']
X = X.values.reshape(-1,1)
Y = Y.values.reshape(-1,1)
```

```python
# Split the data into training sets and testing sets
X_train,X_test = train_test_split(X,test_size=0.2)
Y_train,Y_test = train_test_split(Y,test_size=0.2)

# Create a Linear Regression Ojbect
linear_reg = linear_model.LinearRegression()

# Train the L.R. Model with training sets
linear_reg.fit(X_train,Y_train)

# Make preditions with testing sets
Y_predict = linear_reg.predict(X_test)

# # Show coefficients
# print('Coefficients: ', linear_reg.coef_)

# Mean Squared Error
print('Mean Squared Error: %.2f' % mean_squared_error(Y_test,Y_predict))

# Root Mean Squared Error
print('Root Mean Squared Error: %.2f' % sqrt(mean_squared_error(Y_test, Y_predict)))

# Mean Absolute Error
print('Mean Absolute Error: %.2f' % mean_absolute_error(Y_test,Y_predict))
plt.plot(X_test,Y_predict,'r.')
```
